```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
model = keras.applications.resnet50.ResNet50(weights='imagenet')
```

```python
n_files = 6

images = []
tf_images = []
for i in range (n_files):
    file_path = 'images/file_' + str(i) + '.jpeg'
    image = imread(file_path)
    images.append(image)
    tf_image = tf.image.resize(image, [224, 224])
    tf_images.append(tf_image)

tf_images = tf.stack(tf_images)
tf_images = keras.applications.resnet50.preprocess_input(tf_images)
```

```python
K = 10

Y_proba = model.predict(tf_images)
top_K = keras.applications.resnet50.decode_predictions(Y_proba, top=K)
```

```python
index = 0

for i in range (K):
    name = top_K[index][i][1]
    prob = top_K[index][i][2]
    print('{:15s} {:.2f}%'.format(name, prob*100))

plt.figure(figsize=(6,6))
plt.imshow(images[index])
```
